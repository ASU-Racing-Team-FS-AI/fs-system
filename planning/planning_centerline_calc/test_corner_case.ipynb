```python
import numpy as np
from src.utils.cone_types import ConeTypes
import matplotlib.pyplot as plt
from src.utils.math_utils import angleToVector
from src.corner_case.corner_case_path import CornerCasesPath
```

```python
cones_left = np.array([[4,4],[7,8], [10, 11]])
cones_right = np.array([[5,0],[8,1.5], [10, 2.5]])
carPosition = np.array([3, -2])
carDirection = 0.4
directionvector = angleToVector(carDirection)
cones_input = [np.zeros((0, 2)) for _ in ConeTypes]
cones_input[ConeTypes.BLUE] = cones_left[:0] #specify how many cones are detected
cones_input[ConeTypes.YELLOW] = cones_right[:1]
cornerCasesPath = CornerCasesPath(carPosition, carDirection, cones_input)
path = cornerCasesPath.getPath()
```

```python
plt.scatter(cones_left[:, 0], cones_left[:, 1], c="#7CB9E8", label="left")
plt.scatter(cones_right[:, 0], cones_right[:, 1], c="gold", label="right")
plt.legend()
plt.plot(
    [carPosition[0], carPosition[0] + directionvector[0]],
    [carPosition[1], carPosition[1] + directionvector[1]],
    c="k",
)
plt.scatter(path[:, 0], path[:, 1], c="k", label="path")
```
